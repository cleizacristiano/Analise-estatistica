# TESTE DE CORRELAÇÃO

## Pearson

```python
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregue o arquivo .xlsx
xlsx_file = 'C:/Winpython/A418_H_BOM_JESUS_DA_LAPA_BA.xlsx'
df = pd.read_excel(xlsx_file)
```

```python
#opções de métodos: spearman, kendall ou pearson
df.corr(method = 'pearson')
```

```python
#Caso seja necessário remover alguma variável devido sua baixa correlação
#removendo HORA
corr = df.drop(columns=['HORA']).corr(method = 'pearson')
```

```python
plt.figure(dpi=100)
sns.heatmap(corr, robust=True, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1, fmt='0.2f', linewidths=.5, square=True)
plt.show()
```
