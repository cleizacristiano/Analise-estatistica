# TESTE DE NORMALIDADE

## Q-QPLOT

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregue o arquivo .xlsx
xlsx_file = 'C:/Winpython/A418_H_BOM_JESUS_DA_LAPA_BA_TESTE.xlsx'
df = pd.read_excel(xlsx_file)
```

### 2m TEMPERATURE (t2m)

```python
dados = df['T2M(°C)'].dropna()

plt.figure(figsize=(5, 3.5))  # Largura: 8 polegadas, Altura: 6 polegadas

# Fazer o QQ plot
res = stats.probplot(dados, dist='norm', plot=plt, fit=True)
plt.title('Q-Q Plot - T2M(°C)', fontsize=16)
plt.xlabel('Quantis teóricos', fontsize=14)
plt.ylabel('Valores dos dados', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Personalizar o fundo do gráfico usando o estilo seaborn
sns.set(style='darkgrid')

# Exibir o valor-p no gráfico dentro de uma caixa (annotation box) com cor cinza
plt.annotate(f'p-valor = {p_valor:.4f}', xy=(0.02, 0.98), xycoords='axes fraction', ha='left', va='top',
             bbox=dict(boxstyle="round,pad=0.4", facecolor=(0.9, 0.9, 0.9), edgecolor="gray"), fontsize=12)

plt.show()

# Obter o valor-p (p-valor)
p_valor = res[1][1]


# Definir o nível de significância (0.05 para 5% ou 0.01 para 1%, por exemplo)
nivel_significancia = 0.05

# Verificar se os dados seguem uma distribuição normal
if p_valor < nivel_significancia:
    print("Os dados não seguem uma distribuição normal (p-valor <", nivel_significancia, ")")
else:
    print("Os dados seguem uma distribuição normal (p-valor >=", nivel_significancia, ")")
    
print("O p-valor é:", p_valor)
```
